**知道数据加载的方法后，试试pd.read_csv()和pd.read_table()的不同，如果想让他们效果一样，需要怎么做？了解一下'.tsv'和'.csv'的不同，如何加载这两个数据集？**<br>
read_csv和read_table都是加载带分隔符的数据，每一个分隔符作为一个数据的标志，但二者读出来的数据格式不一样。<br>
read_csv分隔符是逗号(',')，read_table以制表符('\t')作为数据的标志，以行为单位进行存储<br>
如果想让他们格式一样，可以手动设定参数 delimiter = '\t' 或者 delimiter = '，'

tsv - tab-seperated values制表符分割值<br>
csv - comma-seperated values逗号分割值

```python
chunker = pd.read_csv('train.csv', chunksize=500)
```

**【思考】什么是逐块读取？为什么要逐块读取呢？**<br>
逐块读取指将大块的数据分成很多个小块读入<br>
这样可减少内存的存储与计算资源<br>
chunker可以看成是一个迭代器iterator，<br>
An iterator is defined as an object that has an associated next() method that produces consecutive values.

Reference:<br>
Efficient Pandas: Using Chunksize for Large Datasets: https://towardsai.net/p/data-science/efficient-pandas-using-chunksize-for-large-data-sets-c66bf3037f93

```python
type(chunker)
```

```python
for data in chunker:
    print(data)
    print('------one chunk end------')
```
